```python
## this code maps the index value of reduced_g1k15 that matched to swiss cities, creating a dictionary that stores
## all cities matched to the same latitude, longitude values 
##
## defaultdict() help from: https://stackoverflow.com/questions/42137849/defaultdict-first-argument-must-be-callable-or-none
## and https://www.accelebrate.com/blog/using-defaultdict-python/

from collections import defaultdict

cities = defaultdict(lambda:list())

for ind, row in reduced_g1k15.iterrows():
    for elem in indices2:
        if ind == elem[0]:
            if str(elem[0]) in cities:
                cities[str(elem[0])].append(swiss_cities.loc[elem[1]][["City", "Latitude", "Longitude"]])
            else:
                cities[str(elem[0])].append(swiss_cities.loc[elem[1]][["City", "Latitude", "Longitude"]])
        else:
            continue 
```
