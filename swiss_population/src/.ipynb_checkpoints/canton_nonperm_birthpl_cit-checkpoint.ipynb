```python
import pandas as pd
```

```python
canton_nonperm_birthpl_cit = pd.read_csv("../data/2016_canton_perm_birthpl_cit.csv", encoding="latin-1", sep=" ", low_memory=False, skiprows=1)
```

```python
canton_nonperm_birthpl_cit.head()
```

```python
canton_nonperm_birthpl_cit.head()
```

```python
canton_nonperm_Canada_cit = canton_nonperm_birthpl_cit[["Year", "Canton (-) / District (>>) / Commune (......)", "Population type", 
                            "Place of birth", "Citizenship - Total", "Switzerland", "Canada"]]
```

```python
## this block of code divides the column "Canton (-) / District (>>) / Commune (......)" into 3 columns 
##
## loop through each row and assign the canton, district and commune to their respective column

Canton = [] 
District = []
Commune = []

for ind, row in canton_nonperm_Canada_cit[["Canton (-) / District (>>) / Commune (......)"]].iterrows():
    if row[0] == "Switzerland":
        Canton.append("")
        District.append("")
        Commune.append("")
    elif row[0][0] == "-":
        Canton.append(row[0][2:])
        District.append("")
        Commune.append("")
        canton_row_before = row[0][2:]
    else:
        Canton == Canton.append(canton_row_before)
        if row[0][0] == ">":
            District.append(row[0][3:])
            Commune.append("")
            district_row_before = row[0][3:]
        else:
            District == District.append(district_row_before)
            if row[0][0] == ".":
                Commune.append(row[0][11:])
                comm_row_before = row[0][11:]
            else:
                Commune == Commune.append(comm_row_before)
```

```python
canton_nonperm_Canada_cit["Canton"] = Canton
canton_nonperm_Canada_cit["District"] = District
canton_nonperm_Canada_cit["Commune"] = Commune
```

```python
canton_nonperm_Canada_cit = canton_nonperm_Canada_cit[["Year", "Commune", "District", "Canton", "Canton (-) / District (>>) / Commune (......)", 
                           "Population type", "Place of birth", "Citizenship - Total", "Switzerland", "Canada"]]
```
